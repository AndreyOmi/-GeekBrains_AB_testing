## Задача №1:
Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. \
Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. \
Дайте краткие рекомендации команде. \
Результаты: \
1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 \
2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента:  n1 = 164 и n2 = 228 \
3) Конверсии : conv1 =  1.05%  conv2 =  1.47%. 

```python
from scipy.stats import chi2_contingency
```

```python
# создание таблицы сопряженности для входных данных
data = [[164, 228], [(15550-164), (15550-228)]]
data
```

```python
# проведение теста хи-квадрат
stat, p, dof, expected = chi2_contingency(data)
```

```python
# просмотр результатов статистического теста
print("Значение статистического критерия: ", stat)
print("р-значение: ", p)
print("Число степеней свободы: ", dof)

# просмотр таблицы с теоретическим соотношением, если бы значения конверсий были бы несвязаны с проведением эксперимента
print("Теоретическая таблица сопряженности:")
print(expected)
```

```python
# интерпретация результатов хи-теста
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print("Нулевая гипотеза о равенстве конверсий отвергается.")
else:
    print("Нулевая гипотеза о равенстве конверсий принимается.")
```

## Ответ: 
Упрощение формы заказа статистически значимо увеличило конверсию в магазине Утконос.

## Задача №2:
Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе. \
Какой статистический критерий тут лучше всего подойдёт и почему?


# Ответ:
Конверсия представляет из себя значения в процентах. Соответственно, для сравнительного анализа конверсий следует использовать статистические тесты для сравнения долей, например, критерий хи-квадрат или двухпропорционный Z-тест. Далее, важно, что из себя представляют эти 10000 элементов в каждой группе: если это значения "купил/не купил", то на осноании них можно как раз рассчитать конверсии в % в контрольной группе (вариант А) и в экспериментальной группе (вариант В), а затем для провести сравнения с помощью критерия хи-кватрат (как в задаче № 1). \
Если же 10000 элементов в каждой группе - это значения конверсий (в %) в контрольной группе (вариант А) и в экспериментальной группе (вариант В), то для каждой пары конверсий можно по-отдельности пременить хи-квадрат и получить 10000 значений p-value, на основе которых построить, например, 95% доверительный интервал для p-значения и, если он пересекает границу в 0,05 , то сделать вывод о незначимости различий в конверсии в группах, а, если этот доверительный интервал не включает р-значение равное 0,05, то сделать вывод о статистической значимости различий в конверсии в контрольной и экспериментальной группе.
